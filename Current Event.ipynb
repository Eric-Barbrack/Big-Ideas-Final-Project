```python
import pandas as pd
import numpy as np
%matplotlib inline
from numpy import random
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("PublicEvents.csv")
frame = ["neighborhood"]
dataFrame = pd.read_csv("PublicEvents.csv", usecols = frame)
dataFrame = dataFrame.dropna()
dataFrame.reset_index(drop=True, inplace=True)
print(dataFrame)
```

```python
hoodDictionary = dict()

for i in range(len(data)):
    hood = str(data.iloc[i][4])
    if(hood in hoodDictionary):
        value = hoodDictionary[hood]
        hoodDictionary[hood] = value + 1
    else:
        hoodDictionary[hood] = 1

print(hoodDictionary)
```

```python
sortedData = dict(sorted(hoodDictionary.items()))

plt.figure(figsize=(20, 10))
plt.bar(sortedData.keys(), sortedData.values())
plt.title("Arrest Quantity vs Neighborhood")
plt.xlabel("Neighborhood")
plt.xticks(rotation=90)
plt.ylabel("Quantity of Arrests")
```

### We decided that the more public events there are in a neighborhood, the better of an area it is. Because of this, Downtown pittsurgh is clearly the best when it comes to this metric. With 38, it beats the strip district with 3.
